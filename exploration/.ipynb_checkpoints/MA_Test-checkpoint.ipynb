```python
import pandas as pd
import plotly.graph_objects as go
import datetime as dt
from plotting import CandlePlot
```

```python
pair = "EUR_USD"
granularity = "H4"
df = pd.read_pickle(f"../data/{pair}_{granularity}.pkl")
MA_LIST = [10, 20, 50, 100, 200]
```

```python
df.columns

```

```python
df_ma = df[['mid_o', 'mid_h', 'mid_l', 'mid_c', "time"]].copy()
```

```python
for ma in MA_LIST:
    df_ma[F'MA_{ma}'] = df_ma.mid_c.rolling(window=ma).mean()
df_ma.dropna(inplace=True)
df_ma.reset_index(inplace=True, drop=True)
```

```python
df_ma.head()
```

```python
df_plot = df_ma.iloc[:500]
```

```python
cp = CandlePlot(df_plot)
# cp.df_plot = df_plot
```

```python
traces = [f"MA_{ma}" for ma in MA_LIST ]
```

```python

cp.show_plot(line_traces=[f"MA_{ma}" for ma in MA_LIST ])
```

```python
MA_S = "MA_10"
MA_L = "MA_20"
BUY = 1
SELL = -1
NONE = 0
```

```python
df_an = df_ma[['mid_o', 'mid_h', 'mid_l', 'mid_c', "time", MA_S, MA_L]].copy()
```

```python
df_an.head(40)
```

```python
df_an["DELTA"] = df_an.MA_10 - df_an.MA_20
df_an["DELTA_PREV"] = df_an.DELTA.shift(1)
```

```python
def is_trade(row):
    if row.DELTA >= 0 and row.DELTA_PREV < 0 :
        return BUY
    elif row.DELTA < 0 and row.DELTA_PREV >= 0 :
        return SELL
    return NONE
```

```python
df_an["TRADE"] = df_an.apply(is_trade, axis=1)
```

```python
df_trades = df_an[df_an.TRADE != NONE]
```

```python
df_trades.head()
```
